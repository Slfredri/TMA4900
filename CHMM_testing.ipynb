```python
import numpy as np
from hmmlearn import hmm

# Generating synthetic data
np.random.seed(42)

# Number of hidden states
n_states = 2

# Length of the sequence
n_samples = 1000

# Generating synthetic measurements
train_measurements = -7 + np.random.randn(n_samples)
car_measurements = -9 + np.random.randn(n_samples)

# Creating an input sequence with a mix of car and train measurements
observations = np.concatenate([train_measurements, car_measurements])

# Reshape the observations to a 2D array (required by hmmlearn)
observations = observations.reshape(-1, 1)

# Creating the CHMM model
model = hmm.GaussianHMM(n_components=n_states, covariance_type="full")

# Fitting the model to the data
model.fit(observations)

# Generating a new sequence for classification
new_measurements = np.array([-5, -7.5, -8.3, -8.8]).reshape(-1, 1)

# Using the model to predict the most likely hidden states (classes)
predicted_states = model.predict(new_measurements)

# Mapping the predicted states to object classes
predicted_classes = ["Train" if state == 0 else "Car" for state in predicted_states]

# Displaying the results
print("Predicted Classes:", predicted_classes)
```
